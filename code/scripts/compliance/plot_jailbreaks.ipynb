```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = ['Llama2 7B Chat', 'Llama2 7B Chat Uncensored', 'Llama2 13B Chat', 'Llama2 13B Chat Uncensored']
templates = ['APOPHIS', 'Better DAN', 'Maximum']

# Number of successful jailbreaks for each combination of model and template
successful_jailbreaks = {
    'Llama2 7B Chat': {
        'APOPHIS': 132/360,
        'Better DAN': 112/360,
        'Maximum': 141/360,
    },
    'Llama2 7B Chat Uncensored': {
        'APOPHIS': 59/360,
        'Better DAN': 250/360,
        'Maximum': 135/360,
    },
    'Llama2 13B Chat Uncensored': {
        'APOPHIS': 258/360,
        'Better DAN': 360/360,
        'Maximum': 250/360,
    },
    'Llama2 13B Chat': {
        'APOPHIS': 114/360,
        'Better DAN': 315/360,
        'Maximum': 44/360,
    }
}

toxic_responses = {
    'Llama2 7B Chat': {
        'APOPHIS': 20/360,
        'Better DAN': 19/360,
        'Maximum': 1/360,
    },
    'Llama2 7B Chat Uncensored': {
        'APOPHIS': 0,
        'Better DAN': 12/360,
        'Maximum': 0,
    },
    'Llama2 13B Chat Uncensored': {
        'APOPHIS': 7/360,
        'Better DAN': 210/360,
        'Maximum': 0,
    },
    'Llama2 13B Chat': {
        'APOPHIS': 1/360,
        'Better DAN': 24/360,
        'Maximum': 0,
    }
}

successful_roleplays = {
    'Llama2 7B Chat': {
        'APOPHIS': 218/360,
        'Better DAN': 190/360,
        'Maximum': 235/360,
    },
    'Llama2 7B Chat Uncensored': {
        'APOPHIS': 80/360,
        'Better DAN': 360/360,
        'Maximum': 226/360,
    },
    'Llama2 13B Chat Uncensored': {
        'APOPHIS': 360/360,
        'Better DAN': 360/360,
        'Maximum': 360/360,
    },
    'Llama2 13B Chat': {
        'APOPHIS': 114/360,
        'Better DAN': 315/360,
        'Maximum': 44/360,
    }
}



def plot_data(data,type_name):
    # Plotting
    fig, ax = plt.subplots()

    # Set the bar width
    bar_width = 0.2  # Adjusted bar width for better separation

    # Position of bars on X-axis
    r_positions = np.arange(len(templates))

    # Plotting the bars
    for i, model in enumerate(models):
        ax.bar(r_positions + i * bar_width, [data[model][template] for template in templates], width=bar_width, label=model)

    # Adding labels
    ax.set_xlabel('Jailbreak Template')
    ax.set_ylabel(f'{type_name} rate')
    ax.set_title(f'Comparison of {type_name} for Llama2 LLMs')
    ax.set_xticks([r + bar_width for r in r_positions])
    ax.set_xticklabels(templates)
    ax.legend()

    # Show the plot
    plt.show()
    
plot_data(successful_jailbreaks, 'Successful Jailbreaks') 
plot_data(successful_roleplays, 'Successful Role-Plays')
plot_data(toxic_responses, 'Toxic Responses')
```
